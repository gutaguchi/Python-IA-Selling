```python
import pandas as pd
import numpy as np
```

```python
arquivo = pd.read_csv('/Users/gustavotaguchi/Desktop/ifood/ml_project1_data.csv', 
            usecols=['Year_Birth', 'Education', 'Marital_Status', 'Income', 'Kidhome'
                     , 'Teenhome', 'MntWines', 'MntFruits', 'MntMeatProducts'
                     , 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds'
                     , 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases'
                     , 'NumStorePurchases'])
```

```python
arquivo.head()
```

```python
arquivo = arquivo.fillna(arquivo.mean())
```

```python
arquivo['Education'] = arquivo['Education'].replace('Graduation', 1)
arquivo['Education'] = arquivo['Education'].replace('PhD', 2)
arquivo['Education'] = arquivo['Education'].replace('Basic', 3)
arquivo['Education'] = arquivo['Education'].replace('Master', 4)
arquivo['Education'] = arquivo['Education'].replace('2n Cycle', 5)

arquivo['Marital_Status'] = arquivo['Marital_Status'].replace('Single', 1)
arquivo['Marital_Status'] = arquivo['Marital_Status'].replace('Together', 2)
arquivo['Marital_Status'] = arquivo['Marital_Status'].replace('Widow', 3)
arquivo['Marital_Status'] = arquivo['Marital_Status'].replace('Married', 4)
arquivo['Marital_Status'] = arquivo['Marital_Status'].replace('Divorced', 5)
arquivo['Marital_Status'] = arquivo['Marital_Status'].replace('Absurd', 6)
arquivo['Marital_Status'] = arquivo['Marital_Status'].replace('YOLO', 7)
arquivo['Marital_Status'] = arquivo['Marital_Status'].replace('Alone', 7)
```

```python
arquivo.head()
```

```python
#Wine
arquivo.loc[(arquivo['MntWines'] > arquivo['MntFruits']) & (arquivo['MntWines'] > arquivo['MntMeatProducts']) &
            (arquivo['MntWines'] > arquivo['MntFishProducts']) & (arquivo['MntWines'] > arquivo['MntSweetProducts']) &
            (arquivo['MntWines'] > arquivo['MntGoldProds'])
            , 'resultado'] = 0
```

```python
#Fuits
arquivo.loc[(arquivo['MntFruits'] > arquivo['MntWines']) & (arquivo['MntFruits'] > arquivo['MntMeatProducts']) &
            (arquivo['MntFruits'] > arquivo['MntFishProducts']) & (arquivo['MntFruits'] > arquivo['MntSweetProducts']) &
            (arquivo['MntFruits'] > arquivo['MntGoldProds'])
            , 'resultado'] = 1
```

```python
#Meat
arquivo.loc[(arquivo['MntMeatProducts'] > arquivo['MntWines']) & (arquivo['MntMeatProducts'] > arquivo['MntFruits']) &
            (arquivo['MntMeatProducts'] > arquivo['MntFishProducts']) & (arquivo['MntMeatProducts'] > arquivo['MntSweetProducts']) &
            (arquivo['MntMeatProducts'] > arquivo['MntGoldProds'])
            , 'resultado'] = 2
```

```python
#Fish
arquivo.loc[(arquivo['MntFishProducts'] > arquivo['MntWines']) & (arquivo['MntFishProducts'] > arquivo['MntFruits']) &
            (arquivo['MntFishProducts'] > arquivo['MntMeatProducts']) & (arquivo['MntFishProducts'] > arquivo['MntSweetProducts']) &
            (arquivo['MntFishProducts'] > arquivo['MntGoldProds'])
            , 'resultado'] = 3
```

```python
#Sweet
arquivo.loc[(arquivo['MntSweetProducts'] > arquivo['MntWines']) & (arquivo['MntSweetProducts'] > arquivo['MntFruits']) &
            (arquivo['MntSweetProducts'] > arquivo['MntMeatProducts']) & (arquivo['MntSweetProducts'] > arquivo['MntFishProducts']) &
            (arquivo['MntSweetProducts'] > arquivo['MntGoldProds'])
            , 'resultado'] = 4
```

```python
#Fish
arquivo.loc[(arquivo['MntGoldProds'] > arquivo['MntWines']) & (arquivo['MntGoldProds'] > arquivo['MntFruits']) &
            (arquivo['MntGoldProds'] > arquivo['MntMeatProducts']) & (arquivo['MntGoldProds'] > arquivo['MntFishProducts']) &
            (arquivo['MntGoldProds'] > arquivo['MntFishProducts'])
            , 'resultado'] = 5
```

```python
arquivo.head()
```

```python
arquivo['resultado'] = arquivo['resultado'].apply(np.int64)
```

```python
#Separando as variaveis entre preditoras e variavel alvo
y = arquivo['resultado']
x = arquivo.drop('resultado', axis = 1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
#criando os conjuntos de dados de treino e teste - 30% da base de dados pra testar e 70% p treinar
#4 retornos
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3)
```

```python
x_treino.shape
```

```python
type(y_teste)
```

```python
#Algoritmo de machine learning

from sklearn.ensemble import ExtraTreesClassifier
#Criacao modelo
modelo = ExtraTreesClassifier(n_estimators=100)
modelo.fit(x_treino, y_treino)

#imprimindo resultado
#Compara os dados treinados com dados de testes
resultado = modelo.score(x_teste, y_teste)
print("Acurácia:", resultado)
```
